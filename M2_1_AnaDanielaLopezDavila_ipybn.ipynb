```python
#Columna 'Ticket'
ticket = data['Ticket'].unique()
print('Cantidad de valores diferentes en la columna "Ticket": ', len(ticket))
print()
ticket
```

```python
#Columna 'Cabin'
cabin = data['Cabin'].unique()
print('Cantidad de valores diferentes en la columna "Cabin": ', len(cabin))
print()
cabin
```

```python
#Columna 'Embarked'
embarked = data['Embarked'].unique()
print('Cantidad de valores diferentes en la columna "Embarked": ', len(embarked))
print()
embarked
```
